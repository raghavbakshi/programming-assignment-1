```python
#1. Write a Python program to find words which are greater than given length k?
def func1(a,x):
    list1=[]
    for i in a:
        if len(i)>x:
            list1.append(i)
    return list1
```

```python
func1(['dfs','dgf','kiuyu','ew','ouiyu'],3)
```

```python
#2. Write a Python program for removing i-th character from a string?
def func2(a,x):
    for i in a:
        if i!=x:
            print(i,end='')            
```

```python
func2('raghav','a')
```

```python
#4. Write a Python to check if a given string is binary string or not?
string=set(input('enter the string--->'))
set1={'0','1'}
if (string==set1 or string=={'1'} or string=={'0'}):
    print('binary string')
else:
    print('not a binary string')

 
```

```python
#5.Write a Python program to find uncommon words from two Strings?
a=input('enter string 1 ----->')
b=input('enter string 2 ----->')
x=a.split()
y=b.split()
list1=[]
for i in x:
    if i not in y:
        list1.append(i)
for j in y:
    if j not in x:
        list1.append(j)
list1
```

```python
#6. Write a Python to find all duplicate characters in string?
a=input('enter string 1 ----->')
b=input('enter string 2 ----->')
x=a.split()
y=b.split()
list1=[]
for i in x:
    if i in y:
        list1.append(i)
for j in y:
    if j in x:
        list1.append(j)
set(list1)
```

```python
#7. Write a Python Program to check if a string contains any special character?
a=input('enter a string---->')
special_character=('[@_!#$%^&*()<>?/\|}{~:]')
flag=0
for i in a:
    if i in special_character:
        flag=1
if flag==1:
    print('string contains any special character')
else:
    print('string does not contains any special character')
```
