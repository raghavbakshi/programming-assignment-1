```python
#Question 1 Create a function that takes a number as an argument and returns True or False depending on whether the number is symmetrical or not.

def is_symmetrical(n):
    a=''.join(reversed(str(n)))
    a=int(a)
    if n==a:
        return True
    else:
        return False
```

```python
#Question 2 Given a string of numbers separated by a comma and space, return the product of the numbers.

def multiply_nums(string):
    string=string.split(',')
    result=1
    for i in string:
        result*=int(i)
    return result
    
```

```python
multiply_nums('1,2,3,4')
```

```python
#Question 3 Create a function that squares every digit of a number.

def square_digits(num):
    s=''
    for i in str(num):
        s+=str(int(i)*int(i))
    return int(s)
```

```python
square_digits(9119)
```

```python
#Question 4 Create a function that sorts a list and removes all duplicate items from it.

def setify(lst):
    return list(set(lst))
```

```python
setify([3, 3, 3, 2, 1])
```

```python
#Question 5 Create a function that returns the mean of all digits.

def mean(n):
    mean=0
    n=str(n)
    l=len(n)
    for i in n:
        mean+=int(i)
    return mean/l
    
```

```python
mean(12345) 
```
