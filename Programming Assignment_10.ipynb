```python
#1. Write a Python program to find sum of elements in list?
def sum_list(a):
    if type(a[0])==str:
        sum1=''
    else:
        sum1=0
    for i in a:
        sum1=sum1+i
    return sum1
```

```python
sum_list([1,2,3,4,5,6,7,8,9])
```

```python
sum_list(['dfs','dgf','kiuyu','ew','ouiyu'])
```

```python
#2. Write a Python program to Multiply all numbers in the list?
def multiply(x):
    result=1
    for i in x:
        result=result*i
    return result
```

```python
multiply([1,2,3,4,5,6,7,8,9])
```

```python
#3. Write a Python program to find smallest number in a list?
def small_num(a):
    a.sort()
    return a[0]
    
```

```python
small_num([56,49,21,645,148,45,478,34,85,41,255])
```

```python
#4. Write a Python program to find largest number in a list?
def largest_num(a):
    a.sort()
    return a[-1]
```

```python
largest_num([56,49,21,645,148,45,478,34,85,41,255])
```

```python
#5. Write a Python program to find second largest number in a list?
def second_largest_num(a):
    a.sort()
    return a[-2]
```

```python
second_largest_num([56,49,21,645,148,45,478,34,85,41,255])
```

```python
#6. Write a Python program to find N largest elements from a list?
def n_largest_num(a,i):
    a.sort()
    return a[-1:-i-1:-1]
```

```python
#7. Write a Python program to print even numbers in a list?
def even_number(a):
    even=[i for i in a if i%2==0]
    return even
    
```

```python
even_number([56,49,21,645,148,45,478,34,85,41,255])
```

```python
#8. Write a Python program to print odd numbers in a List?
def odd_number(a):
    odd=[i for i in a if i%2!=0]
    return odd
```

```python
odd_number([56,49,21,645,148,45,478,34,85,41,255])
```

```python
9. Write a Python program to Remove empty List from List?
def remove_empty_list(a):
    lst=[]
    for i in a:
        if i!=[]:
            lst.append(i)
    return lst
```

```python
remove_empty_list([56,[],[45,562],[],'fd',[],['fds','wfd']])
```

```python
#10. Write a Python program to Cloning or Copying a list?
def cloning(x):
    a=x.copy()
    return a
```

```python
cloning([49, 21, 645, 45, 85, 41, 255])
```

```python
#11. Write a Python program to Count occurrences of an element in a list?
def count(a,j):
    count1=0
    for i in a:
        if i==j:
            count1+=1
    return count1
```

```python
count([49, 21, 645, 45, 21,85, 41, 255,21],21)
```
