```python
#1. Write a Python program to Extract Unique values dictionary values?
dict1={'a':1,'b':2,'c':3,'d':'raghav','e':1,'f':'raghav'}
lst=[]
for i in dict1.values():
    if i not in lst:
        lst.append(i)
lst
```

```python
#6. Write a Python program to check order of character in string using OrderedDict()?
from collections import OrderedDict
def func(dic1,pattern):
    dic=OrderedDict.fromkeys(dic1)
    ptrord=0
    for key,value in dic.items():
        if key==pattern[ptrord]:
            ptrord+=1
        if ptrord==len(pattern):
            return True
    return False
    
```

```python
func('raghav','av')
```
