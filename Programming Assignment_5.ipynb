```python
#1. Write a Python Program to Find LCM?
def lcm(number1,number2):
    if number1>number2:
        greater=number1
    else:
        greater=number2
    while(True):
        if ((greater%number1==0) and (greater%number2==0)):
            lcm=greater
            break
        greater+=1
    return lcm
    
```

```python
lcm(12,20)
```

```python
#2.Write a Python Program to Find HCF?
def hcf(x,y):
    if x>y:
        num=x
    else:
        num=y
    while(True):
        if  x%num==0 and y%num==0:
            hcf=num
            break
        num-=1
    return hcf
```

```python
hcf(36,60)
```

```python
# 3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
def deci_to_binary(a):
    if a>1:
        deci_to_binary(a//2)
    print(a%2,end='')
def deci_to_octal(a):
    if a>7:
        deci_to_octal(a//8)
    print(a%8,end='') 
def deci_to_hexadecimal(a):
    return hex(a)
```

```python
deci_to_binary(13)
```

```python
deci_to_octal(52)
```

```python
deci_to_hexadecimal(501)
```
