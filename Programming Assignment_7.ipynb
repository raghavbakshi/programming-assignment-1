```python
#1. Write a Python Program to find sum of array?
array=[1,23,6,4,86,2,4,86,15,2,46,2]
for i in array:
    if type(i)==str:
        sum1=''
    else:
        sum1=0
for i in array:
    sum1=sum1+i
print(sum1)
```

```python
#2.Write a Python Program to find largest element in an array?
array=['fds','fdvdxf','fgvx','kyu','kyuguf','ewret']
for i in array:
    if type(i)==str:
        x=''
    else:
        x=0
for i in array:
    if i>x:
        x=i
print(x)
        
```

```python
#3. Write a Python Program for array rotation?
array=[1,23,6,4,86,2,86,15,2,46,2]
a=int(input('rotate array with----->'))
lst=[]
for i in array:
    if a!=i:
        lst.append(i)
        
    else:
        break
lst.append(a)
for i in lst:
    array.remove(i)

for i in lst:
    array.append(i)
array
```

```python
#4. Write a Python Program to Split the array and add the first part to the end?
array=[1,23,6,4,86,2,86,15,2,46,2]
a=int(input('rotate array with----->'))
lst=[]
for i in range(len(array)):
    if a!=i:
        lst.append(array[i])
        
    else:
        break

for i in lst:
    array.remove(i)

for i in lst:
    array.append(i)
array

```

```python
#5. Write a Python Program to check if given array is Monotonic?
array=[1,2,3,4,4,4,4,5,6,6,6,6,7,7,7,8,9]
a=sorted(array)
b=sorted(array,reverse=True)
if (array==a or array==b):
    print('array is Monotonic')
else:
    print('array is not Monotonic')
```
