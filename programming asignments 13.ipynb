```python
#Question 1:Write a program that calculates and prints the value according to the given formula:
#Q = Square root of [(2 * C * D)/H]
#Following are the fixed values of C and H:
#C is 50. H is 30.
#D is the variable whose values should be input to your program in a comma-separated sequence.

C=50
H=30
D=(input("enter the number in comma seperated sequence--->"))

lst=[]
for i in D.split(','):
    i=int(i)
    Q=pow((2*C*i/H),1/2)
    Q=int(Q)
    lst.append(Q)
lst
```

```python
#Question 2:Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The
#element value in the i-th row and j-th column of the array should be i*j.
#Note: i=0,1.., X-1; j=0,1,Â¡Y-1.

import numpy as np
X=int(input('enter x---->'))
Y=int(input('enter y---->'))
lst=[]
for i in range(X):
    list1=[]
    for j in range(Y):
        list1.append(i*j)
    lst.append(list1)
print(lst)
```

```python
#Question 3:Write a program that accepts a comma separated sequence of words as input and prints the
#words in a comma-separated sequence after sorting them alphabetically.

words=input('enter the words in comma seperated form---->')
word=words.split(',')
word.sort()
print(word)  
```

```python
#Question 4:Write a program that accepts a sequence of whitespace separated words as input and prints
#the words after removing all duplicate words and sorting them alphanumerically.

words=input('enter the words in whitespace seperated form---->')
word=words.split()
word.sort()
list2=[]
for i in word:
    if i not in list2:
        list2.append(i)
list2
```

```python
#Question 5:Write a program that accepts a sentence and calculate the number of letters and digits.
sentence=input('enter a sentence--->')
count=0
count1=0
for i in sentence:
    if i.isalpha():
        count+=1
    elif i.isdigit():
        count1+=1
print('the number of alphabets are ', count)
print('the number of digits are',count1)
```
