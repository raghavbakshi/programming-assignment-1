```python
#1. Write a Python Program to Find the Factorial of a Number?

def fact(n):
    for i in range((n-1),0,-1):
        n=n*i
    return n
```

```python
#2. Write a Python Program to Display the multiplication Table?
for i in range(2,4):
    for j in range(1,11):
        print(i,'*',j,'=',i*j)
    print('\r')
```

```python
#3. Write a Python Program to Print the Fibonacci sequence?
n=int(input('length of fibonacci series required---->'))
sum1=0
a=0
b=1
count=1
while count<=n:
    print(sum1)
    count+=1
    a=b
    b=sum1
    sum1=a+b
```

```python
#4. Write a Python Program to Check Armstrong Number?
a=int(input('enter a number---->'))
order=len(str(a))
num=a
sum1=0
while num>0:
    digit=num%10
    sum1=sum1+pow(digit,order)
    num=num//10
    
if(sum1==a):
    print('arms')
else:
    print('not')
```

```python
#5. Write a Python Program to Find Armstrong Number in an Interval?
a=int(input('start value---->'))
b=int(input('end value----->'))
for i in range(a,b):
    order=len(str(i))
    num=i
    sum1=0
    while num>0:
        digit=num%10
        sum1=sum1+pow(digit,order)
        num=num//10
    if(sum1==i):
        print(i)
```
