```python
#4. Write a Python program to solve quadratic equation?
a=int(input('enter a___'))
b=int(input('enter b___'))
c=int(input('enter c___'))
d=(pow(b,2)-4*a*c)
x1=(-b+pow(d,1/2))/(2*a)
x2=(-b-pow(d,1/2))/(2*a)
print('1st root of the equatin is',x1)
print('2ns root of equation is',x2)
```
